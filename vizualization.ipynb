# Data Visualization with Matplotlib and Seaborn

Welcome to the workshop on data visualization using Matplotlib and Seaborn! Today, we'll be working with the Iris dataset to create various types of plots. 

## Introduction

The Iris dataset is a classic dataset in the field of machine learning and statistics. It includes 150 observations of iris flowers with four features: sepal length, sepal width, petal length, and petal width, along with the species of each flower.

Let's start by importing the necessary libraries and loading the dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Mapping target numbers to species names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
iris_df['species'] = iris_df['species'].map(species_map)

# Display the first few rows of the dataset
iris_df.head()
```

#### Scatter Plot with Matplotlib

A scatter plot displays values for two variables for a set of data. 

To create a scatter plot with Matplotlib, use the `plt.scatter()` function. Here's how you can create a scatter plot for 'sepal length' and 'sepal width':

```python
# Scatter plot for sepal length and sepal width
plt.figure(figsize=(10, 6))
plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c='blue')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.show()
```

#### Line Plot with Matplotlib

A line plot is useful for displaying trends over time or ordered data points. 

To create a line plot, use the `plt.plot()` function. Here's how to create a line plot for 'sepal length':

```python
# Line plot for sepal length
plt.figure(figsize=(10, 6))
plt.plot(iris_df['sepal length (cm)'], c='b')
plt.title('Sepal Length Trend')
plt.xlabel('Index')
plt.ylabel('Sepal Length (cm)')
plt.show()
plt.figure(figsize=(10, 6))
plt.plot(iris_df['sepal width (cm)'], c='g')
plt.title('Sepal width Trend')
plt.xlabel('Index')
plt.ylabel('Sepal width (cm)')
plt.grid(True)
plt.show()
```

#### Histogram with Matplotlib

A histogram displays the distribution of a continuous variable by dividing the data into bins.

To create a histogram, use the `plt.hist()` function. Here's how to create a histogram for 'petal length':

```python
# Histogram for petal length
plt.figure(figsize=(10, 6))
plt.hist(iris_df['petal length (cm)'], bins=150, color='purple')
plt.title('Distribution of Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

#### Box Plot with Matplotlib

A box plot shows the distribution of a variable through its quartiles.

To create a box plot, use the `plt.boxplot()` function. Here's how to create a box plot for 'sepal width':

```python
# Box plot for sepal width
plt.figure(figsize=(10, 6))
plt.boxplot(iris_df['sepal width (cm)'])
plt.title('Box Plot of Sepal Width')
plt.ylabel('Sepal Width (cm)')
plt.grid(True)
plt.show()
```

## What is Correlation?

Correlation measures the strength and direction of a linear relationship between two variables. The correlation coefficient (often denoted as `r`) ranges from -1 to 1.

- **Positive correlation**: As one variable increases, the other variable also increases.
- **Negative correlation**: As one variable increases, the other variable decreases.
- **No correlation**: There is no linear relationship between the variables.

### Types of Correlation Coefficients
- **Pearson's correlation coefficient**: Measures linear correlation between two variables.
- **Spearman's rank correlation coefficient**: Measures the rank-order correlation.

#### Heatmap with Seaborn

A heatmap displays the correlation between different variables in a dataset.

To create a heatmap, use the `sns.heatmap()` function. Here's how to create a heatmap for the Iris dataset: (use irid_df.corr to find correlations)

```python
plt.figure(figsize=(10, 6))
sns.heatmap(iris_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Iris Features')
plt.show()
```

#### Violin Plot with Seaborn

A violin plot shows the distribution of the data across different categories.

To create a violin plot, use the `sns.violinplot()` function. Here's how to create a violin plot for 'sepal length' across different species:

```python
# Violin plot for sepal length across species
plt.figure(figsize=(10, 6))
sns.violinplot(x='species', y='sepal length (cm)', data=iris_df, palette='muted')
plt.title('Violin Plot of Sepal Length by Species')
plt.show()
```

#### Catplot with Seaborn

A catplot allows us to plot categorical data with several subtypes (e.g., box, violin, bar).

To create a catplot, use the `sns.catplot()` function. Here's how to create a catplot for 'petal length' across different species:

```python
# Catplot for petal length across species
sns.catplot(x='species', y='petal length (cm)', kind='box', data=iris_df, height=6, aspect=1)
plt.title('Catplot of Petal Length by Species')
plt.show()
```

#### Joint Plot with Seaborn

A joint plot shows the bivariate scatter plots and univariate histograms in the same figure.

To create a joint plot, use the `sns.jointplot()` function. Here's how to create a joint plot for 'sepal length' and 'sepal width':

```python
# Joint plot for sepal length and sepal width
sns.jointplot(x='sepal length (cm)', y='sepal width (cm)', data=iris_df, kind='scatter', height=8)
plt.title('Joint Plot of Sepal Length and Sepal Width')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Mapping target numbers to species names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
iris_df['species'] = iris_df['species'].map(species_map)

# Group by species
grouped = iris_df.groupby('species')

# Iterate over groups and create joint plots
for species_name, group_data in grouped:
    # Create joint plot for sepal length vs sepal width
    sns.jointplot(x='sepal length (cm)', y='sepal width (cm)', data=group_data, kind='scatter', height=8)
    plt.title(f'Joint Plot for {species_name}')
    plt.show()
```
